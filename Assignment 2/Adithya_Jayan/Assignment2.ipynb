Answer 1

```
import pandas as pd
from google.colab import files
url='/content/ex1data1.txt'
colnames=['Population', 'Profit'] 
df=pd.read_csv(url,names=colnames)
df
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```
x=df[['Population']]
y=df.Profit
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```
model.fit(x_train,y_train)
```

```
model.score(x_test,y_test)
```

```
y_test
```

```
model.predict(x_test).max()
```

The city from the test set with maximum predicted profit will be:

```
prediction_1 = pd.DataFrame({'city population':y_test}).reset_index(drop=True)
prediction_2 = pd.DataFrame({'Predicted Profit':model.predict(x_test)}).reset_index(drop=True)
prediction = pd.concat([prediction_1, prediction_2], axis=1)
prediction.max()
```

Answer 2

```
url='/content/ex2data1.txt'
colnames=['Test1','Test2','Result'] 
df=pd.read_csv(url,names=colnames)
df
```

```
Inputs=df[['Test1','Test2']]
Target =df.Result
Inputs_train, Inputs_test, Target_train, Target_test=train_test_split(Inputs,Target,test_size=0.2)
```

```
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(Inputs_train,Target_train)
```

```
model.score(Inputs_test,Target_test)
```

```
model.predict(Inputs_test)
```

```
Final_1 = pd.DataFrame({'Actual_Result':Target_test}).reset_index(drop=True)
Final_2 = pd.DataFrame({'Model_Predicted_Result':model.predict(Inputs_test)}).reset_index(drop=True)
Final = pd.concat([Final_1,Final_2], axis=1)
Final
```
