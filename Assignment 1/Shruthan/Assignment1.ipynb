```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("../PL18-19.csv")
```

```python
data.head()
```

```python
# Dictionary to store the new names of the columns
cols = {
    'FTHG' : 'Full Time Home Goals',
    'FTAG' : 'Full Time Away Goals',
    'FTR' : 'Full Time Result',
    'HTHG' : 'Half Time Home Goals',
    'HTAG' : 'Half Time Away Goals',
    'HTR' : 'Half Time Result',
    'HS' : 'Home Shots',
    'AS' : 'Away Shots',
    'HST': 'Home Shot Target',
    'AST': 'Away Shot Target',
    'HF' : 'Home Fouls',
    'AF' : 'Away Fouls',
    'HC' : 'Home Corners',
    'AC' : 'Away Corners',
    'HY' : 'Home Yellow Cards',
    'AY' : 'Away Yellow Cards',
    'HR' : 'Home Red Cards',
    'AR' : 'Away Red Cards'
}
```

```python
# Renaming the columns
data.rename(columns = cols, inplace = True)
```

```python
# Prints all participating teams in lexicographical order 
print(np.union1d(data.HomeTeam, data.AwayTeam))
```

```python
# Prints maximum number of shots in a match
max(data['Home Shot Target'] + data['Away Shot Target'])
```

```python
data[data['Referee'] == 'J Linington']
```

```python
# Prints the number of cards given out by each referee
total_cards = data['Home Yellow Cards'] + data['Away Yellow Cards'] + data['Home Red Cards'] + data['Away Red Cards']
total_cards.groupby(data['Referee']).sum()
```

```python
print(data['Home Yellow Cards'].groupby(data['HomeTeam']).count())
```

```python
# To print teams with maximum cards in the season
HomeYellow = data['Home Yellow Cards'].groupby(data['HomeTeam']).sum()
HomeRed = data['Home Red Cards'].groupby(data['HomeTeam']).sum()
AwayYellow = data['Away Yellow Cards'].groupby(data['AwayTeam']).sum()
AwayRed = data['Away Red Cards'].groupby(data['AwayTeam']).sum()

TotalCards = HomeYellow + HomeRed + AwayYellow + AwayRed

print("The team with maximum cards during the season is: ", end = " ")
print(TotalCards[TotalCards == max(TotalCards)].index[0])
```

```python
# Referee with maximum home wins
HomeWin = data[data['Full Time Result'] == 'H']
HomeWinPerReferee = HomeWin.groupby(HomeWin['Referee'])['Date'].count()
print(HomeWinPerReferee[HomeWinPerReferee == max(HomeWinPerReferee)].index[0])
```

```python
# Half time winner is full time winner
half_and_full_winner = data[data['Half Time Result'] == data['Full Time Result']]
print("Percentage where half time winner is the full time winner =", (half_and_full_winner.size / data.size) * 100)
```

```python
# Ratio of average shots on target  to total shots
data_target = data['Home Shot Target'] + data['Away Shot Target']
data_shot = data['Home Shots'] + data['Away Shots']
average_shots = data_shot.sum()/data.size
average_target = data_target.sum()/data.size
print("Ratio of average shots on target to total shots =", round(average_target/average_shots, 2))
```

```python
winner = []
loser = []
drawn = []
for i in range(len(data)):
    if (data.loc[i, 'Full Time Result'] == 'H'):
        winner.append(data.loc[i, 'HomeTeam'])
        loser.append(data.loc[i, 'AwayTeam'])
        
    elif(data.loc[i, 'Full Time Result'] == 'A'):
        winner.append(data.loc[i, 'AwayTeam'])
        loser.append(data.loc[i, 'HomeTeam'])
    elif(data.loc[i, 'Full Time Result'] == 'D'):
        drawn.append(data.loc[i, 'AwayTeam'])
        drawn.append(data.loc[i, 'HomeTeam'])
```

```python
winner = pd.Series(winner)
loser = pd.Series(loser)
drawn = pd.Series(drawn)
```

```python
winner_table = pd.DataFrame(winner.value_counts(), columns = ['Won'])
loser_table = pd.DataFrame(loser.value_counts(), columns = ['Lost'])
draw_table = pd.DataFrame(drawn.value_counts(), columns = ['Drawn'])
```

```python
winner_table['Lost'] = loser_table
winner_table['Drawn'] = draw_table
```

```python
winner_table
```

```python
winner_table['Played'] = winner_table['Won'] + winner_table['Lost'] + winner_table['Drawn']
```

```python
table = winner_table[['Played', 'Won', 'Drawn', 'Lost']]
```

```python
table
```

```python
table['Points'] = (3 * table['Won']) + table['Drawn']
```

```python
table.sort_values(by = 'Points', inplace = True, ascending = False)
```
