```
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ACM-NITK/Machine-Learning-SMP-2019/master/Assignment%201/PL18-19.csv')
df.head()
```

```
df1=df.rename(index=str,columns={"FTHG" : "Goals by Home at Full","FTAG" : "Goals by Away at Full","HTHG" : "Goals by Home at Half","HTAG" : "Goals by Away at Half", "FTR" : "Full Time Results","HTR" : "Half Time Result"})
```

```
df1=df1.rename(index=str,columns={"HS" : "Home Shots","AS" : "Away Shots","HST" : "Home Shots on Target","AST" : "Away Shots on Target","HF" : "Home fouls","AF" : "Away Fouls","HC" : "Home Corners","AC" : "Away Corners","HY" : "Home Yellow Cards","AY" : "Away Yellow Cards","HR" : "Home Red Cards","AR" : "Away Red Cards"})
```
