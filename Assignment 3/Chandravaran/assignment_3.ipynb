```
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ACM-NITK/Machine-Learning-SMP-2019/master/Assignment%203/titanic.csv')
df.head()
```

```
d={'male': 1,'female' : 0}
df['Sex1']=df['Sex'].map(d)
df.head()
```

```
a=df.index
a
```

```
df['index']=a
df.head()
```

```
df=df.reset_index()
```

```
df=df[(0.0<df['Age'])&(df['Age']<100.0)]
df
```

```
x=df[['Pclass','Sex1','Age']]
y=df.Survived
```

```
from sklearn.tree import DecisionTreeClassifier
r = DecisionTreeClassifier()
```

```
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
r.fit(x_train,y_train)
```

```
scores = list()
for i in range(3):
    x_train, x_test, y_train, y_test= train_test_split(x, y, test_size = 0.25)
    scores.append(r.score(x_test, y_test))
print("The mean score of three different splittings is: %0.2f" % np.mean(scores))
```
