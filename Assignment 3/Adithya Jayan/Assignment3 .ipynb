I have assumed that the NaN in the Cabin column refers to 'Not Having a cabin' and not 'Missing Data'


```
import pandas as pd
url= 'https://raw.githubusercontent.com/ACM-NITK/Machine-Learning-SMP-2019/master/Assignment%203/titanic.csv'
df=pd.read_csv(url)
```

```
df.head()
```

```
df.info()
```

```
dfNew=df.dropna(subset=['Sex', 'Embarked'], how='any')
```

```
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
le_Embarked = LabelEncoder()
```

```
dfNew['Sex']= le_Sex.fit_transform(dfNew['Sex'])
dfNew['Embarked']= le_Embarked.fit_transform(dfNew['Embarked'])
dfNew
```

```
import numpy as np
dfNew=dfNew.fillna(0)
dfNew['Cabin']=np.where((dfNew['Cabin']==0),0,1)
dfNew
```

```
dfNew=dfNew.drop(['Ticket','Fare','Name','PassengerId'],axis='columns')
dfNew
```

```
X=dfNew.drop(['Survived'],axis='columns')
X
```

```
Y=dfNew.Survived
Y
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.25)
```

```
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```
a=model.score(x_test,y_test)
a
```

```
x_train2, x_test2, y_train2, y_test2=train_test_split(X,Y,test_size=0.25)
b=model.score(x_test2,y_test2)
b
```

```
x_train3, x_test3, y_train3, y_test3=train_test_split(X,Y,test_size=0.25)
c=model.score(x_test3,y_test3)
c
```

```
AvgScore=(a+b+c)/3
AvgScore
```
